```python
# Thêm thư viện
from PIL import Image
```

```python
import os
```

```python
# Thư mục hiện tại
print(os.getcwd())
# Thư mục chứa ảnh
directory = 'D:/MonHoc/ThiGiacMayTinh/File/Image'
```

```python
# Tạo đường dẫn đến ảnh
image_path = directory + '/p01.png'
print(image_path)
```

```python
# Sử dụng Image.open() đọc ảnh
img01 = Image.open(image_path)
```

```python
# Dùng display() để hiển thị ảnh
display(img01)
```

```python
# In định dạng của ảnh
print('Định dạng của ảnh là: ', img01.format)
```

```python
# In ra size ảnh
print('Kích thước của ảnh: ', img01.size)
```

```python
# Chuyển đôi định dạng của ảnh
# Chuyển đổi hình ảnh sang chế độ RGB trước khi lưu
img01 = img01.convert("RGB")
new_jpg_image_path = directory + '/p01_NEW.jpg'
img01.save(new_jpg_image_path)
```

```python
# Đóng tập tin
img01.close()
```
