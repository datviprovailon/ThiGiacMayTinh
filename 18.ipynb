```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from PIL import Image
```

```python
# Đường dẫn ảnh
img_path = 'D:/MonHoc/ThiGiacMayTinh/File/Image/p12.png'
img = Image.open(img_path)
```

```python
# Chuyển sang ảnh xám
gray_image = img.convert('L')

# Chuyển ảnh xám về mảng
image_arry = np.array(gray_image)

# Chuyển sang ảnh nhị phân
binary_image = 1*(image_arry<200)

# Đánh nhãn
labeled_image, num_features = ndimage.label(binary_image)
```

```python
# Tạo lưới subplot với 3 dòng và 1 cột
fig, axs = plt.subplots(3, 1, figsize=(50, 50))

plt.subplot(3,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title("Ảnh xám")

plt.subplot(3,1,2)
plt.imshow(binary_image, cmap='gray')
plt.title("Ảnh nhị phân")

plt.subplot(3,1,3)
plt.imshow(labeled_image, cmap='gray')
plt.title(f"Đã đánh nhãn ({num_features} đối tượng)")

plt.show();
```

```python
# Chuyển sang ảnh xám
gray_image = img.convert('L')

# Chuyển ảnh xám về mảng
image_arry = np.array(gray_image)

# Chuyển sang ảnh nhị phân
binary_image = 1*(image_arry<200)

# Thực hiện opening để loại bỏ nhiễu và nối các đối tượng
opened_image = ndimage.binary_opening(binary_image, structure=np.ones((11,11)))

# Đánh nhãn
labeled_image, num_features = ndimage.label(opened_image)

# Tạo lưới subplot với 3 dòng và 1 cột
fig, axs = plt.subplots(4, 1, figsize=(50, 50))

plt.subplot(4,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title("Ảnh xám")

plt.subplot(4,1,2)
plt.imshow(binary_image, cmap='gray')
plt.title("Ảnh nhị phân")
                                      
plt.subplot(4,1,3)
plt.imshow(opened_image, cmap='gray')
plt.title("Ảnh sau opening")


plt.subplot(4,1,4)
plt.imshow(labeled_image, cmap='gray')
plt.title(f"Đã đánh nhãn ({num_features} đối tượng)")

plt.show();
```
