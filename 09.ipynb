```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Tạo đường dẫn lưu ảnh
directory = 'D:/MonHoc/ThiGiacMayTinh/File/Image'
# Đọc ảnh
my_path = 'D:/MonHoc/ThiGiacMayTinh/File/Image/p06.png'
img = Image.open(my_path)
# Hiển thị ảnh
plt.imshow(img)
```

```python
# Chuyển sang ảnh xám
img_gray = img.convert('L')
plt.imshow(img_gray, cmap='gray')
```

```python
img_array = np.array(img_gray)
print(img_array)
```

```python
# Tạo một hình vẽ mới
plt.figure()
plt.hist(img_array.flatten(), bins=128)
plt.title('Biểu đồ Histogram của ảnh')
plt.xlabel('Giá trị của pixel')
plt.ylabel('Số lượng')
plt.show()
```

```python
# Làm nổi bật đường biên
plt.figure()
plt.gray()
plt.contour(img_gray, origin="image")
plt.show()
```

```python
# Đường dẫn ảnh mới
my_path_02 = 'D:/MonHoc/ThiGiacMayTinh/File/Image/p07.png'

# Đọc ảnh và hiển thị ảnh với trục tọa độ
img_02 = Image.open(my_path_02).convert("L")

# Hiển thị ảnh trên trục tọa độ
plt.imshow(img_02)

# Tạo một ình vẽ mới
plt.figure()

# Đặt màu sắc là xám
plt.gray()

# Vẽ biểu đồ đường biên (contour) của ảnh
plt.contour(img_02, origin="image")

# Đảm bảo tỷ lệ trục x và trục y trên biểu đồ là bằng nhau
plt.axis('equal')

# hiển thị ảnh
plt.show()
```
