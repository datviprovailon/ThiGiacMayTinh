```python
# Đọc ảnh
my_path = 'D:/MonHoc/ThiGiacMayTinh/File/Image/p03.png'
img = load_image(my_path)
```

```python
# Chuyển từ ảnh RGB sang ảnh xám
gray_image = img.convert("L")
display(gray_image)
```

```python
# Kiểm tra kiểu ảnh
if img.mode != 'RGB':
    img = img.convert('RGB')
# Phân RGB thành 3 tầng màu khác nhau
red_band, green_band, blue_band = img.split()
# Hiển thị các kênh màu
display(red_band)
display(green_band)
display(blue_band)
```

```python
# Trộn ảnh
merged_image =  Image.merge("RGB", (red_band, green_band, blue_band))
display(merged_image)
```
