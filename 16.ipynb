```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
# Đường dẫn ảnh
img_path = 'D:/MonHoc/ThiGiacMayTinh/File/Image/p06.png'
img = Image.open(img_path)
```

```python
# Tạo ảnh xám
gray_image = img.convert('L')
```

```python
# Chuyển ảnh xám về mảng
img_a = np.array(gray_image)
```

```python
# Làm mờ ảnh xám với bộ lọc Gaussian
blurred_img_a = ndimage.gaussian_filter(img_a, sigma=1, mode="constant")
blurred_img_b = ndimage.gaussian_filter(img_a, sigma=2, mode="constant")
blurred_img_c = ndimage.gaussian_filter(img_a, sigma=5, mode="constant")
```

```python
# Hiển thị ảnh
fig, axs = plt.subplots(2,2, figsize=(10, 8))

plt.subplot(2,2,1)
plt.imshow(gray_image, cmap='gray')
plt.title('Ảnh xám gốc')

plt.subplot(2,2,2)
plt.imshow(blurred_img_a, cmap='gray')
plt.title('sigma 1')

plt.subplot(2,2,3)
plt.imshow(blurred_img_b, cmap='gray')
plt.title('sigma 3')

plt.subplot(2,2,4)
plt.imshow(blurred_img_c, cmap='gray')
plt.title('sigma 5')
```
