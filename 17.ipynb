```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
# Đường dẫn ảnh
img_path = 'D:/MonHoc/ThiGiacMayTinh/File/Image/p06.png'
img = Image.open(img_path)
```

```python
# Ảnh xám
gray_image = img.convert('L')

# Chuyển ảnh xám về mảng
img_a = np.array(gray_image)
```

```python
# Tính gradient theo hướng x
gradient_x = ndimage.sobel(img_a, axis=0, mode='constant')
```

```python
# Hiển thị ảnh
fig, axs = plt.subplots(2,1, figsize=(10, 8))

plt.subplot(2,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title('Ảnh xám gốc')

plt.subplot(2,1,2)
plt.imshow(gradient_x, cmap='gray')
plt.title('Ảnh gradient_x')

plt.show()
```

```python
# Tính gradient theo hướng x
gradient_x = ndimage.sobel(img_a, axis=1, mode='constant')

# Hiển thị ảnh
fig, axs = plt.subplots(2,1, figsize=(10, 8))

plt.subplot(2,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title('Ảnh xám gốc')

plt.subplot(2,1,2)
plt.imshow(gradient_x, cmap='gray')
plt.title('Ảnh gradient_x')

plt.show()
```

```python
# Tính gradient theo hướng x (đạo hàm riêng theo x) bằng mặt nạ Prewitt x
gradient_x = ndimage.convolve(img_a, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))

# Tính gradient theo hướng y (đạo hàm riêng theo y) bằng mặt nạ Prewitt y
gradient_y = ndimage.convolve(img_a, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))

# Tính biên độ gradient
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

# Hiển thị ảnh
fig, axs = plt.subplots(2,1, figsize=(10, 8))

plt.subplot(2,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title('Ảnh xám gốc')

plt.subplot(2,1,2)
plt.imshow(gradient_magnitude, cmap='gray')
plt.title('Ảnh gradient_x')

plt.show()
```
