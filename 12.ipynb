```python
# Thm th vin
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# ng dn nh
my_path = 'D:/MonHoc/ThiGiacMayTinh/File/Image/p06.png'

# c nh
img = Image.open(my_path)
plt.imshow(img)
```

```python
# Chuyn hnh v array
img_array = np.array(img)

# Kim tra kch thc ca anh
print(img_array.shape)

# Chuyn i sang nh xm
img_array_2 = np.array(img.convert('L'), 'f')
print(img_array_2.shape)
print(img_array_2.dtype)
# Hien thi anh
plt.imshow(img_array_2)
```

```python
# Chuyn i hnh nh thnh mng NumPy
image_array = np.array(img.convert("L"))
```

```python
# Bin i nh xm theo yu cu
inverted_image = 255 - image_array  # Invert nh bng cch ly 255 - gi tr pixel
squared_image = image_array ** 2  # Bnh phng gi tr pixel
clamped_image = np.clip(image_array, 100, 200)  # Gii hn gi tr pixel trong khong 100-200
```

```python
# Hin th hnh nh gc v hnh nh  bin i vi khong trng
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
fig.subplots_adjust(wspace=0.2, hspace=0.2)

axs[0, 0].imshow(image_array, cmap="gray")
axs[0, 0].set_title('Hnh nh gc')

axs[0, 1].imshow(inverted_image, cmap="gray")
axs[0, 1].set_title('Hnh nh  bin i')

axs[1, 0].imshow(squared_image, cmap="gray")
axs[1, 0].set_title('Hnh nh  bin i')

axs[1, 1].imshow(clamped_image, cmap="gray")
axs[1, 1].set_title('Hnh nh  bin i')

plt.show()
```
